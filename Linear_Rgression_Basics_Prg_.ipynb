```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

hours = np.array([[1],[2],[3],[4],[5]])
marks=np.array([30,35,40,45,50])
print(hours.shape)
print(marks.shape)

model = LinearRegression()
model.fit(hours,marks)

predict= model.predict([[8]])

print(f"If the Student study for 8 hours he score mark of: {round(predict[0])}")
```

```python
pred= model.predict(hours)
plt.figure(figsize=(8,5))
plt.scatter(hours,marks, color='blue',label = "Actual Marks")
plt.plot(hours,pred,color='red',label="Predicted Marks")
plt.scatter(8, predict, color='green', label='Prediction for 8 hrs', marker='o', s=100)
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)
plt.show()
```
